Import the libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#### EXTRACT THE SUBCATEGORIES LINKS IN FAQ WEB PAGE

```python
# define the url 
url = "https://kim.metu.edu.tr/en/student-clubs-and-societies"
```

```python
# define the request 
session_req = requests.session()
response  = session_req.get(url)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
paragraphs = soup.find_all("p")
```

```python
paragraphs
```

```python
# Initialize a list to store club details
clubs = []
current_club = None

# Loop through each paragraph
for p in paragraphs:
    # Check if the paragraph contains a club name (based on anchor tags and uppercase naming convention)
    if p.find("a") or p.text.isupper():
        # If there's an existing club being built, append it to the list
        if current_club:
            clubs.append(current_club)
        
        # Start a new club entry
        current_club = {
            "name": p.get_text(strip=True),
            "information": ""
        }
    elif current_club:
        # Accumulate information for the current club
        current_club["information"] += p.get_text(separator=" ", strip=True) + " "

# Append the last club if it exists
if current_club:
    clubs.append(current_club)
```

```python
df = pd.DataFrame(clubs)
```
